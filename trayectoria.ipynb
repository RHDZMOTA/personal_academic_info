### Empty Dataframe 

Construct empty dataframe. 

```python
# index of hours (discrete)
hours = [
    '0700 - 0900', 
    '0900 - 1100',
    '1100 - 1300',
    '1300 - 1500',
    '1500 - 1600',
    '1600 - 1800',
    '1800 - 2000',
    '2000 - 2200'
]

# create inner datastructure for the dataframe 
cols = ['Lunes','Martes','Miércoles','Jueves','Viernes','Sábado','Domingo']
courses = {
    cols[0]:['']*len(hours),
    cols[1]:['']*len(hours),
    cols[2]:['']*len(hours),
    cols[3]:['']*len(hours),
    cols[4]:['']*len(hours),
    cols[5]:['']*len(hours),
    cols[6]:['']*len(hours)
}


# create empty dataframe
empty_schedule = pd.DataFrame(courses, index = hours)
empty_schedule[cols]
```

### Function to build schedule

Create function. 

```python


def build_schedule(select_courses, schedule, return_conflicts = False, print_warnings = True):
    '''
    Function to build and return a schedule...
    '''
    # total number of credits
    total_credits = 0
    for i in select_courses:
        total_credits += i['Credits']
    
    
    # conflict variable 
    conflict = {}
    conflict_id = 0

    # fill dataframe
    for i in select_courses:
        _name = i['Name']
        _days = i['Days']
        _hours = i['Hours']

        # set version zero
        version = 0

        # acces to a set of days and a set of hours
        for h, d in zip(_hours, _days):

            # save into dataframe 
            for sh, sd in zip(h,d):

                # determine real name 
                real_name = '{} V{}'.format(_name, version)

                # check if the course can be saved into dataframe
                if schedule[sd][sh] == '':
                    schedule[sd][sh] = real_name

                # if a previous conflict has been found, append the real_name
                elif 'conflict' in schedule[sd][sh]:
                    conflict[schedule[sd][sh]] = conflict[schedule[sd][sh]].append(real_name)

                # if it is a new conflict, save and register it. 
                else:
                    if print_warnings:
                        print('\nWarning: A conflict has been found. See variable _conflicts_.\n\n')
                    conflict_name = 'conflict_{}'.format(conflict_id)
                    conflict[conflict_name] = [schedule[sd][sh], real_name]
                    schedule[sd][sh] = conflict_name
                    conflict_id += 1

            # update version
            version += 1


    if len(conflict.keys()) != 0:
        print('Conflicts found: \n')
        for i,j in zip(conflict.keys(),conflict.values()):
            string = 'ID: {} :::::: Courses: {}'.format(i,j)
            print(string)
            
    print('\n\nTotal number of credits selected: {}\n\n'.format(total_credits))
            
    if return_conflicts:
        return schedule, conflicts
    
    return schedule
        
```
